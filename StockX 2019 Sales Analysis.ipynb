# Import Data From Excel


```python
import pandas as pd

df = pd.read_csv("StockxData.csv")

print(df)
```

# Change data types 

```python
df['Sale Price'] = df['Sale Price'].astype(int)
print(df.dtypes) 
```

```python
df['Retail Price'] = df['Retail Price'].astype(int)
print(df.dtypes) 
```

# Create new column "Profit" 

```python
df['Profit'] = df['Sale Price'] - df['Sale Price'] * 0.10 - df['Retail Price']
print(df)
```

# Add a new column "Profit" to the data frame

```python
# add a new column 'Profit' 
df = df.assign(Profit=df['Sale Price'] - df['Sale Price'] * 0.10 - df['Retail Price'])

print(df)
```

#  Identify the top ten most profitable sneakers

```python
# group the dataframe by the 'Sneaker Name' column
grouped_df = df.groupby('Sneaker Name')

# calculate the sum of the 'Profit' column for each group
sum_df = grouped_df['Profit'].sum()

sum_df = sum_df.sort_values(ascending=False)

sum_df = sum_df.reset_index()

topten = sum_df.head(10)

print("Top ten most profitable sneakers sold on StockX in 2019\n")

print(topten)




```

# Identify the top ten most sold sneakers and their average profit per sale

```python
# Group the data by the "Sneaker Name" column and calculate the mean of the "Sale Price" and "Profit" columns
sneaker_groups = df.groupby("Sneaker Name")["Sale Price", "Profit"].mean()

# Count the number of occurrences of each value in the "Sneaker Name" column
sneaker_counts = df["Sneaker Name"].value_counts()

# Add the sneaker count to the dataframe
sneaker_groups["Quantity Sold"] = sneaker_counts

# Rename the "Profit" column to "Average Profit"
sneaker_groups = sneaker_groups.rename(columns={"Profit": "Average Profit"})


# Sort the grouped data by the number of occurrences of each sneaker
sorted_sneaker_groups = sneaker_groups.sort_values(by="Quantity Sold", ascending=False)

print("Top ten most sold sneakers 2019 \n")

# Print the results
print(sorted_sneaker_groups.head(10))












```
